```python
split_data = food_df['nutrition'].str.strip('[]').str.split(',', expand=True)

# Renombrar las columnas
split_data.columns = ['calories', 'total fat (PDV)', 'sugar (PDV)', 'sodium (PDV)', 'protein (PDV)', 'saturated fat (PDV)', 'carbohydrates (PDV)']

# Convertir las columnas a tipo numérico
split_data = split_data.apply(pd.to_numeric, errors='coerce')

# Concatenar las nuevas columnas al DataFrame original
food_df = pd.concat([food_df, split_data], axis=1)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
from sklearn.preprocessing import StandardScaler

# Select features for clustering
clustering_data = food_df[['calories', 'total fat (PDV)', 'sugar (PDV)', 'sodium (PDV)', 'protein (PDV)', 'saturated fat (PDV)', 'carbohydrates (PDV)']]
# Scale the data
scaler = StandardScaler()
clustering_data_standardized = scaler.fit_transform(clustering_data)

# Experiment with different hyperparameters
num_clusters_range = range(2, 3)  # Explore a range of cluster numbers
best_silhouette_score = -1
best_num_clusters = None
best_labels = None

for num_clusters in num_clusters_range:
    kmeans = KMeans(n_clusters=num_clusters, init='k-means++', n_init=10, max_iter=100, random_state=42)
    labels = kmeans.fit_predict(clustering_data_standardized)

    # Calculate evaluation metrics
    silhouette_avg = silhouette_score(clustering_data_standardized, labels)
    calinski_harabasz = calinski_harabasz_score(clustering_data_standardized, labels)
    davies_bouldin = davies_bouldin_score(clustering_data_standardized, labels)

    print(f"Number of clusters: {num_clusters}")
    print(f"Silhouette score: {silhouette_avg:.4f}")
    print(f"Calinski-Harabasz score: {calinski_harabasz:.4f}")
    print(f"Davies-Bouldin score: {davies_bouldin:.4f}")

    # Track the best scores and labels
    if silhouette_avg > best_silhouette_score:
        best_silhouette_score = silhouette_avg
        best_num_clusters = num_clusters
        best_labels = labels

# Visualize the results (optional)
# You can create visualizations to further explore the clusters

print("\nBest results:")
print(f"Number of clusters: {best_num_clusters}")
print(f"Silhouette score: {best_silhouette_score:.4f}")

# Use the best model for further analysis
best_kmeans = KMeans(n_clusters=best_num_clusters, init='k-means++', n_init=10, max_iter=300, random_state=42)
best_kmeans.fit(clustering_data_standardized)
```

```python
def preprocess_data(df):
    df[['calories', 'total fat (PDV)', 'sugar (PDV)', 'sodium (PDV)', 'protein (PDV)', 'saturated fat (PDV)', 'carbohydrates (PDV)']] = df['nutrition'].str.split(",", expand=True)
    df['calories'] = df['calories'].apply(lambda x: x.replace('[',''))
    df['carbohydrates (PDV)'] = df['carbohydrates (PDV)'].apply(lambda x: x.replace(']',''))
    df[['calories', 'total fat (PDV)', 'sugar (PDV)', 'sodium (PDV)', 'protein (PDV)', 'saturated fat (PDV)', 'carbohydrates (PDV)']] = df[['calories', 'total fat (PDV)', 'sugar (PDV)', 'sodium (PDV)', 'protein (PDV)', 'saturated fat (PDV)', 'carbohydrates (PDV)']].astype('float')
    
    df.drop(['id', 'contributor_id', 'submitted', 'tags', 'nutrition'], axis=1, inplace=True)

    
    df['food types'] = np.nan
    df['food types'] = df['food types'].astype('str')
    
    for i in df.index:
        if 'eggs' not in df.loc[i, 'ingredients']:
            if 'ice-cream' in df.loc[i, 'ingredients'] or 'chocolate' in df.loc[i, 'ingredients'] or 'cookies' in df.loc[i, 'ingredients']:
                df.loc[i, 'food types'] = 'Veg dessert'
        elif 'eggs' in df.loc[i, 'ingredients']:
            if 'ice-cream' in df.loc[i, 'ingredients'] or 'chocolate' in df.loc[i, 'ingredients'] or 'cookies' in df.loc[i, 'ingredients']:
                df.loc[i, 'food types'] = 'Non-Veg dessert'
    
    for i in df.index:
        if df.loc[i, 'food types'] != 'Veg dessert' and df.loc[i, 'food types'] != 'Non-Veg dessert' and 20 < df.loc[i, 'calories'] < 300:
            df.loc[i, 'food types'] = 'Healthy'
    
    for i in df.index:
        if df.loc[i, 'food types'] != 'Veg dessert' and df.loc[i, 'food types'] != 'Non-Veg dessert' and df.loc[i, 'food types'] != 'Healthy':
            if 'chicken' in df.loc[i, 'ingredients'] or 'eggs' in df.loc[i, 'ingredients'] or 'ham' in df.loc[i, 'ingredients'] or 'pepperoni' in df.loc[i, 'ingredients']:
                df.loc[i, 'food types'] = 'Non-veg'
    
    for i in df.index:
        if df.loc[i, 'food types'] != 'Veg dessert' and df.loc[i, 'food types'] != 'Non-Veg dessert' and df.loc[i, 'food types'] != 'Healthy' and df.loc[i, 'food types'] != 'Non-veg':
            df.loc[i, 'food types'] = 'Veg'
    
    df['food types'].value_counts()
    df['food types'].isnull().sum()
    
    return df

# Apply the function to your dataframe
food_df = preprocess_data(food_df)
food_df.head()

# One-hot encode food types
types = pd.get_dummies(food_df['food types'])
food_df = pd.concat([food_df, types], axis=1)

# Create a pivot table for nutrient details
cal = food_df.pivot_table(columns='name', values=['total fat (PDV)', 'sugar (PDV)', 'sodium (PDV)', 'saturated fat (PDV)', 'carbohydrates (PDV)'])

# Apply K-Means clustering
columns_for_clustering = ['calories', 'total fat (PDV)', 'sugar (PDV)', 'sodium (PDV)', 'protein (PDV)', 'saturated fat (PDV)', 'carbohydrates (PDV)']
clustering_data = food_df[columns_for_clustering]
scaler = StandardScaler()
clustering_data_standardized = scaler.fit_transform(clustering_data)

kmeans = KMeans(n_clusters=2000, n_init=10, random_state=42)
food_df['cluster'] = kmeans.fit_predict(clustering_data_standardized)
food_df.head(5)
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Standardize the data
scaler = StandardScaler()
clustering_data_standardized = scaler.fit_transform(clustering_data)

# Run K-Means for a range of cluster numbers
max_clusters = 20
inertia_values = []

for k in range(1, max_clusters + 1):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(clustering_data_standardized)
    inertia_values.append(kmeans.inertia_)

# Plot the elbow curve
plt.figure(figsize=(10, 6))
plt.plot(range(1, max_clusters + 1), inertia_values, marker='o')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Sum of Squared Distances (Inertia)')
plt.show()
```

```python
import seaborn as sns

# Perform PCA for 2D visualization
pca = PCA(n_components=2)
clustering_data_pca = pca.fit_transform(clustering_data_standardized)

# Add the PCA components to the DataFrame
food_df['PCA1'] = clustering_data_pca[:, 0]
food_df['PCA2'] = clustering_data_pca[:, 1]

# Count the number of clusters
num_clusters = len(food_df['cluster'].unique())
print(f"Total number of clusters: {num_clusters}")

# Display some statistics for each cluster
cluster_stats = food_df.groupby('cluster').agg({'calories': ['mean', 'std', 'count']})
print("\nCluster Statistics:")
print(cluster_stats)

# Create a 2D scatter plot
plt.figure(figsize=(10, 8))
sns.scatterplot(x='PCA1', y='PCA2', hue='cluster', data=food_df, palette='viridis', alpha=0.7)

# Annotate some values in each cluster
for cluster in range(num_clusters):
    cluster_center = kmeans.cluster_centers_[cluster]
    plt.annotate(f'Cluster {cluster}\nMean Calories: {cluster_stats.loc[cluster, ("calories", "mean")]:.2f}\nCount: {cluster_stats.loc[cluster, ("calories", "count")]}',
                 xy=(cluster_center[0], cluster_center[1]),
                 xytext=(cluster_center[0] + 1, cluster_center[1] + 1),
                 arrowprops=dict(facecolor='black', shrink=0.05))

plt.title('K-Means Clustering Visualization (2D) with Annotations')
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.legend(title='Cluster')
plt.show()
```

```python
def plot_pie_chart(ax, data, labels):
    ax.pie(data, labels=labels, autopct='%1.1f%%', startangle=90, colors=['#99ff99', '#ffcc99', '#ccffcc', '#ff6666', '#c2f0c2'])
    ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    ax.set_facecolor('#1a1a1a')
# Define the function to find the closest food items to a target
def find_closest_food(target_food, target_features, exclude_foods=[]):
    recommended_items = []
    target_cluster_df = food_df.loc[food_df['name'] == target_food, 'cluster']
    if target_cluster_df.empty:
        print(f"Target food {target_food} not found in the dataset.")
        return recommended_items
    target_cluster = target_cluster_df.values[0]

    # Filter food items in the same cluster, excluding those already shown
    cluster_recipes = food_df[food_df['cluster'] == target_cluster]
    cluster_recipes = cluster_recipes[~cluster_recipes['name'].isin(exclude_foods)]

    # Check if any items remain after exclusion
    if cluster_recipes.empty:
        print(f"No recipes found in the same cluster as {target_food} after excluding previous recommendations.")
        return recommended_items

    # Calculate distances to the target
    cluster_data = clustering_data_standardized[food_df['cluster'] == target_cluster]
    distances = np.linalg.norm(cluster_recipes[columns_for_clustering] - target_features, axis=1)
    cluster_recipes.loc[:, 'distance_to_target'] = distances

    # Sort recipes by distance and find the closest ones
    closest_recipes = cluster_recipes.sort_values(by='distance_to_target').head(5)  # Consider top 5 recommendations

    recommended_items.append(closest_recipes)

    return recommended_items

# Meal recommendations
cal1 = {
    'Breakfast': ['say what   banana sandwich'],
    'Mid_Morning_Snack': ['sour cream  avocado dip  vegan'],
    'Lunch': ['now and later  vegetarian empanadas'],
    'Evening_Snack': ['the man s  taco dip'],
    'Dinner': ['mexican pasta']
}
selected_foods = []
recommended_items = []
lose_weight = 1
# Loop through meal recommendations
for meal_time, foods in cal1.items():


    for food in foods:
        target_features = food_df[food_df['name'] == food][columns_for_clustering].values[0]
        closest_food_items = find_closest_food(food, target_features, exclude_foods=selected_foods)
        selected_foods.extend(closest_food_items[0]['name'].tolist())
        recommended_items.append(closest_food_items)

    print(f"\n{meal_time} Recommendations for {foods}:")

    option_number = 1
    for rec_item in recommended_items:
        print(f"\nClosest 3 Items to {foods}:")

        if not rec_item:
            print(f"\nOption {option_number}: No recommendations found.")
        else:
            if lose_weight == 1:
                # If losing weight, show the item with the lowest calorie value from the top 5 recommendations
                best_recommendation = rec_item[0].sort_values(by='calories').head(1)
            else:
                # If not losing weight, show the item with the maximum calorie value from the top 10 recommendations
                best_recommendation = rec_item[0].sort_values(by='calories', ascending=False).head(1)
            for _, row in best_recommendation.iterrows():  # No need for enumerate here
                print(f"\nOption {option_number}:")

                print(f"Name: {row['name']}")
                print(f"Minutes: {row['minutes']}")  # Assuming 'minutes' column exists
                print(f"Number of Steps: {row['n_steps']}")  # Assuming 'n_steps' column exists
                print(f"Steps: {eval(row['steps'])[0]}")  # Convert steps from string to list
                print(f"Description: {row['description']}")
                print(f"Ingredients: {row['ingredients']}")
                print(f"Number of Ingredients: {row['n_ingredients']}")  # Assuming 'n_ingredients' column exists
                print(f"Calories: {row['calories']}")
                print(f"Total Fat (PDV): {row['total fat (PDV)']}")
                print(f"Sugar (PDV): {row['sugar (PDV)']}")
                print(f"Sodium (PDV): {row['sodium (PDV)']}")  # Assuming 'sodium (PDV)' column exists
                print(f"Protein (PDV): {row['protein (PDV)']}")  # Assuming 'protein (PDV)' column exists
                print(f"Saturated Fat (PDV): {row['saturated fat (PDV)']}")  # Assuming 'saturated fat (PDV)' column exists
                print(f"Carbohydrates (PDV): {row['carbohydrates (PDV)']}")  # Assuming 'carbohydrates (PDV)' column exists
                
                fig, ax = plt.subplots()
                nutritional_data = [row.loc['total fat (PDV)'], row.loc['sugar (PDV)'], row.loc['sodium (PDV)'], row.loc['protein (PDV)'], row.loc['carbohydrates (PDV)']]
                nutritional_labels = ['Total Fat', 'Sugar', 'Sodium', 'Protein', 'Carbohydrates']
                plot_pie_chart(ax, nutritional_data, nutritional_labels)

                # Display the pie chart
                plt.title(f"Nutritional Information for {row['name']}")
                plt.show()
                plt.show()

                option_number += 1
print("\nEnd of Recommendations")
```
